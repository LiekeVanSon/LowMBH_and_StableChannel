# Plotting $\zeta_{RL}$ as a function of $q$ and $\beta$


## Following Soberman 1997

Their definitions:
\begin{align}
m_d     & \equiv m_2 \\
m_a     & \equiv m_1 \\
q       &\equiv \frac{m_2}{m_1} = \frac{m_d}{m_a} \\
\beta_s & \equiv \frac{-\partial m_{iso-re} }{\partial m_d}
\end{align}
$\partial m_{iso-re}$ is mass isotropically reemitted from the accretor, but I have no clue what the signs do...??
Pols defines the signs differently.. In any case, I'm pretty sure that 

$$
\beta_p = 1-\beta_s
$$
or 
$$
\beta_s = 1 - \beta_p
$$

***
***


We start with equation 62:

The definition of $\zeta_{RL}$ can be rewritten with the product rule:

\begin{align}
\zeta_{RL} & \equiv \frac{\partial \ln r_L}{\partial \ln m_d} \\
           & = \color{red}{\underbrace{ \frac{\partial \ln a}{\partial \ln m_d} }_{(1)} }\
           + \color{green}{\underbrace{\left( \frac{\partial \ln(r_L/a) }{\partial \ln q} \right) }_{(2)} }
           \cdot \color{blue}{\underbrace{\left( \frac{\partial \ln q}{\partial \ln m_d} \right) }_{(3)} }         
\end{align}




***
***

$\large \color{red}{Part (1)}$


Straight from eq. 16.8 in Polls
using that $\partial \ln a = \frac{\partial a}{a} $, and that $\gamma = m_d/m_a$ for isotropic reemission:

\begin{align}
\color{red}{\frac{\partial \ln a}{\partial \ln m_d} }  
& = -2 \left[ 1 - \beta_p \frac{m_d}{m_a} - \left(1 - \beta_p \right)\left(\frac{m_d}{m_a}  + 1/2 \right) \frac{m_d}{(m_d + m_a)} \right]\\
& = \underbrace{ -2 \left[ 1 - \beta_p q - \left(1 - \beta_p \right)\left(q  + \frac{1}{2} \right) \frac{q}{q + 1} \right] }_{(1)} \\
\end{align}

I think we may actually want:

\begin{align}
\frac{a_f}{a_i} = \left( \frac{q_i}{q_f}  \right)^{2} 
                   \left( \frac{1+q_i}{1+q_f}  \right)
                   \left[ \frac{1+ \beta_p q_f}{1 + \beta_p q_i } \right]^{5 + \frac{2(1-\beta_p) }{\beta_p}} 
\end{align}

With:

$$
\begin{align}
q_i   &= m_{d,i}/m_{a,i}  &
q_f   &= m_{d,f}/m_{a,f}  
\end{align}
$$

#### For the first mass transfer, we have

$$
\begin{align}
q_f   &= \frac{f_{core,1} \cdot M_{ZAMS,1} }{M_{ZAMS,2} + M_{ZAMS,1}\cdot \beta_p(1-f_{core}) }  
       = \frac{f_{core,1}}{q_{ZAMS} + \beta_p(1-f_{core,1}) } \\
q_i   &= \frac{M_{ZAMS,1}}{M_{ZAMS,2}} = \frac{1}{q_{ZAMS}} \\
\end{align}
$$



```python
def part1(q, beta):
    """    
    dln(a)/dln m_d
    note q == m_donor/m_accretor
    """
    return -2 * (1 - beta*q - (1 - beta) * (q + 1./2.) * (q / (q + 1) )  )
    
    
    
```

***
***



$\large \color{green}{Part (2)}$

We know from Eggleton:

$$
\frac{r_L}{a} = \frac{0.49 q^{2/3}}{0.6q^{2/3} + \ln(1 + q^{1/3}) }
$$

If you derive this (I'm just believing Soberman here for a bit

\begin{align}
\color{green}{\left( \frac{\partial \ln(r_L/a) }{\partial \ln q} \right)}  & =
\underbrace{\frac{q^{1/3}}{3} \cdot \left( \frac{2}{q^{1/3}} - \frac{ 1.2q^{1/3} + \frac{1}{1+q^{1/3}} }{ 0.6 q^{2/3} + \ln(1 + q^{1/3})  }       \right)}_{(2)} 
\end{align}



```python
def part2(q):
    """
    dln(Rl/a)/dlnq
    note q == m_donor/m_accretor
    """
    A = (q**(1./3.)) / 3.
    B = 2./(q**(1./3.))
    C = (  1.2 * q**(1./3.) +  1./(1. + q**(1./3)) ) / (0.6 * q**(2./3.) + np.log(1 + q**(1./3.) ) )
    return A * (B - C)
    
    
    
```


***
***



$\large \color{blue}{Part (3)}$

(from eq. 12 in Soberman)
we neglect wind mass loss, so $\alpha = 0$

\begin{align}
\color{blue}{\left( \frac{\partial \ln q}{\partial \ln m_d} \right) }         
& = (1 + \epsilon_w q) \\
& = 1 + \left(1 - \alpha - \beta_s  \right) q \\
& = 1 + \left(1 - (1 - \beta_p) \right) q \\
& = 1 + \left(1 -1 + \beta_p \right) q \\
& = \underbrace{  1 +\beta_p q }_{(3)} \\
\end{align}




```python
def part3(q, beta):
    """
    dln(q)/dln m
    note q == m_donor/m_accretor
    """
    return 1 + beta*q
    
    
    
```

### Aweseome! Let's bring all three parts together...

```python
np.seterr(divide='ignore')
np.seterr(invalid='ignore')

q_list    = np.arange(0,5, 0.1)
betas     = np.arange(0,1.25,0.25)
colors    = ['#c32204', '#f95f21','#ffa37e','#ffc8b5', '#450b01'] #oranges
# sns.color_palette('rocket', len(betas))

###########
fig, ax = plt.subplots(figsize = (10,10))

for i, B in enumerate(betas):
    print(colors[i])
    zeta_rl = part1(q_list, B) + (part2(q_list) * part3(q_list, B))
    plt.plot(q_list, zeta_rl, label = r'$%s$'%(np.round(B,2) ), color = colors[i])


###########
# indicator lines
ax.axhline(y = 0, ls = '--', color = 'grey', lw = 3) #

ax.axhline(y = 2., ls = '-', color = 'grey', lw = 2) #default zeta MS
ax.text(0.5, 2.1, '$\zeta_{MS}$', color = 'grey') #default zeta MS

for zeta_hg in [3.5,4.5,5.5,6.5]:
    ax.axhline(y = zeta_hg, ls = ':', color = 'grey', lw = 1) #variations in zeta HG

ax.axhline(y = 6.0, ls = '-', color = 'grey', lw = 3) #default zeta HG
ax.text(0.5, 6.1, '$\zeta_{eff}$', color = 'k', clip_on = True) #default zeta MS




###########
# Plot values
plt.legend(title = r'$\beta_{\rm{acc}}$', loc = 'lower right')
plt.xlabel(r'$q = m_{donor}/m_{accretor}$', fontsize = 30)
plt.ylabel(r'$\zeta_{RL}$', fontsize = 35)

plt.xlim(0, 5)
plt.ylim(-1, 7)

# increase number of ticks
ax.xaxis.set_ticks_position('both')
ax.xaxis.set_minor_locator(matplotlib.ticker.AutoMinorLocator(5))

ax.yaxis.set_ticks_position('both')
ax.yaxis.set_minor_locator(matplotlib.ticker.AutoMinorLocator(5))

# plt.grid()

plt.savefig(proj_dir + '/plots/zeta_RL.pdf')
plt.show()

```
